```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Let x1 = x and x2 = y
f = lambda x, y: 4*x + 6*y - 2*((x)**2) - 2*x*y - 2*((y)**2)
```

```python
x = np.linspace(-5,10,100)  # range of values of x
y = np.linspace(-5,10,100)  # range of values of y

X,Y = np.meshgrid(x,y)      #creating a meshgrid
F = f(X,Y)                  #saving the function values with respect to corresponding values of x and y

figure = plt.figure(figsize=(10,10))    #specifying the size of the figure
ax = plt.subplot(projection='3d')   # specifying that the graph should be 3-dimensional
graph = ax.plot_surface(X,Y,F, cmap= 'gist_rainbow')    # plotting the graph
figure.colorbar(graph)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
plt.show()
```
