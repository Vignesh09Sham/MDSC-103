```python
import numpy as np
import matplotlib.pyplot as plt
```

### Gradient Descent

```python
f = lambda x: x**2
delta = (1 * 10**(-5))
xp = 4
xn = 0
x = np.linspace(-5, 5, 100)

list = [xp]
for i in range(0,100):
    xn = xp - 0.1 * (2 * xp)
    list.append(xn)     #appending all the x values generated by the gradient descent formula to the list
    if(not((xn - xp) < delta or (xn - xp) > -delta)):   #come out of loop if condition is false
        break
    xp = xn
    
print(xn)
grad_descent_y= [f(i) for i in list]
plt.plot(x, f(x))    #plotting the graph y = x^2
plt.title("Gradient Descent for y = x^2")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.plot(list, grad_descent_y, color='red', marker='o')     #plotting the gradient descent

plt.legend(["graph: y = x^2", "Gradient Descent"])
plt.show()
```

```python
def GD_Function(func, grad_func, x0, learning_rate, delta, lower_range_limit, upper_range_limit):
    f = func
    g = grad_func
    x = np.linspace(lower_range_limit, upper_range_limit,100)
    xn = 0
    list = [x0]
    for i in range(0,100):
        xn = x0 - (learning_rate * g(x0))
        list.append(xn)     #appending all the x values generated by the gradient descent formula to the list
        if(not((xn - x0) < delta or (xn - x0) > -delta)):   #come out of loop if condition is false
            break
        x0 = xn

    print(xn)

    grad_descent_y = [f(i) for i in list]
    plt.plot(x, f(x))       #plotting the given graph
    plt.title("Gradient Descent for y = x^3")
    plt.xlabel("x-values")
    plt.ylabel("y-values")
    plt.plot(list, grad_descent_y, color='red', marker='o')     #plotting the gradient descent

    plt.legend(["graph: y = x^3", "Gradient Descent"]) 
    plt.show()
```

```python
GD_Function(lambda x: x**3, lambda x: 3*(x**2), 2, 0.09, (1 * 10**(-5)), -5,5)
```
