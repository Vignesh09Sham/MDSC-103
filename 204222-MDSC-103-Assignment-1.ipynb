 $Minimize f(x,y) = 6x^2+ 3y^2+ 4xy$


subject to: $xy = 7$

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
```

```python
f = lambda x: 6*(x[0]**2) + 3*(x[1]**2) + 4*x[0]*x[1]
X0 = (5,5)
cons = ({'type': 'eq', 'fun': lambda x: x[0]*x[1] - 7})

res1 = minimize(f, X0, constraints=cons)
res2 = minimize(f,X0)
min1 = res1.x
min2 = res2.x
print(min1)
print(min2)
res2
```

```python
func = lambda x,y: 6*(x**2) + 3*(y**2) + 4*x*y
x = np.linspace(-5,10,100)
y = np.linspace(-5,10,100)

X,Y = np.meshgrid(x,y)
F = func(X,Y)

fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection='3d') # set up the axis as a 3D plot
ax.scatter3D(min1[0], min1[1],func(min1[0],min1[1]),c="black",marker="*",s=100)
ax.scatter3D(min2[0], min2[1],func(min2[0],min2[1]),c="black",marker="o",s=100)
surf = ax.plot_surface(X,Y,F, cmap= 'gist_rainbow')
fig.colorbar(surf)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('func(x,y)')
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-5, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((100,100))

fig = plt.figure(figsize=(10,10))

levels = np.arange(0,12,0.5)
cp = plt.contour(X, Y, Z,levels=levels, cmap='gist_rainbow')
plt.clabel(cp)
#plt.plot(x, 7/x, 'k:', linewidth=1)
plt.plot(min1[0],min1[1],c='r',marker='*')
plt.plot(min2[0], min2[1], c = 'b',marker='o')
plt.show()
```
